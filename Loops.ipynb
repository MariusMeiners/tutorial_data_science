# Loops

## What types of loops exist in Python?

### while loop

```python
# This is pseudo code -> dont run this
while andrew.isUpset:
    andrew.scream()
```

conceptionally: run the loop as long as the condition is met  
-> used if the NUMBER of runs is unknown before looping  
Here we do not know when andrew will stop being upset so we let him scream until he calms down

```python
#EXAMPLE
# Imagine the index method did not exist
api_response = [1, 2, 3, 4, 5, 6, 7]
wanted_number = 5
found = False
index = 0

while found == False:
    if api_response[index] == wanted_number:
        found = True
    else:
        index = index + 1

print("wanted number was found at : {}".format(index))
```

#### special while loop -> infinite loop

```python
while 1 == 1:
    tradingAPI.observeStockMarket()
```

used when you want your program do run in a contionous loop  
examples:  
video game: you always want to render the next image.  
trading algorithm: always check the current stock market data and make trading decisions

### examples for loop

```python
# convention: when possible name lists in a plural way (numbers)
# notice: we know the number of iterations beforehand (length of numbers -> for loop)
numbers = [1, 2, 3, 4, 5, 6, 7]
x = 10
x_added = []
for number in numbers:
    print(number)
    x_added.append(number + x)
print(x_added)
```

```python
# while loop for comparison
# this is more complicated since we have to define the condition ourselves and remember to update the index
numbers = [1, 2, 3, 4, 5, 6, 7]
x = 10
x_added = []

index = 0
length_of_list = len(numbers)

while index < length_of_list:
    print(numbers[index])
    x_added.append(numbers[index] + x)
    index = index + 1
    
print(x_added)
```

### understanding the in operator

```python
# use to check if list CONTAINS a specific element
numbers = [1, 2, 3, 4, 5, 6, 7]
searched_number = 4

if searched_number in numbers:
    print("the number {} was found in the list of numbers".format(searched_number))
```

### looping with index

```python
# imagine people standing in a line infront of a concert
line = ["John", "Maya", "Brian", "Mark", "Lisa", "Jane"]
# you want to know which person is standing at which position

for person in line:
    #this is terrible because it wastes a huge amount of computing power
    position_of_person_in_line = line.index(person)
    print("Person {} is standing at position {}".format(person, position_of_person_in_line))
    
print("\n")
# way better:
for index, person in enumerate(line):
    print("Person {} is standing at position {}".format(person, index))
```

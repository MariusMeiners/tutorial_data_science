# Loops

## What types of loops exist in Python?

### While loop

```python
# This is pseudo code -> dont run this
while andrew.isUpset:
    andrew.scream()
```

conceptionally: run the loop as long as the condition is met  
-> used if the NUMBER of runs is unknown before looping  
Here we do not know when andrew will stop being upset so we let him scream until he calms down

#### special while loop -> infinite loop

```python
while 1 == 1:
    tradingAPI.observeStockMarket()
```

used when you want your program do run in a contionous loop  
examples:  
video game: you always want to render the next image.  
trading algorithm: always check the current stock market data and make trading decisions 

### For loop

```python
for i in range(10):
    print(i)
```
